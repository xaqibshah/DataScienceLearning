# 03- How to rename the columns

```python
df
df = pd.DataFrame({'A col': [1,2,3,4,5,6,7,8,9], 'B col':[10,11,12,13,14,15,16,18,19], 'C col':[21,22,23,24,25,26,27,28,29]})
df
```

```python
df.rename(columns={'A col': 'col_a', 'B col': 'col_b'}, inplace=True)
df.rename(columns={'C col': 'col_c'}, inplace=True)
df
```

```python
df.columns=['Col_aa', 'Col_bb', 'col_cc']
df
df.columns=df.columns.str.replace('_', '-')
df
df.columns=df.columns.str.replace('-', '*')
df
```

```python
# adding prefix and sufex to columns

df.add_prefix('unit-')
df.add_suffix('-ten')
df
```
