# 9- convert string to number

```python
df = pd.DataFrame({'col=a': [2,4,6,8],
'col=b': [1,3,5,7]})

df.dtypes
```

```python
# convert integer to string
# df = df.astype(str)
# another way to convert
df = df.astype({'col=a': 'float64', 'col=b': 'str'})

df.dtypes
```

```python
#convert string to integer
# df=df.astype(int)
# another way for converting
pd.to_numeric(df['col=a'], errors='coerce')


```
