# 16- Spliting a string into multiple columns

```python
df = pd.DataFrame({'name': ['ahmed Raza', 'Ali Nawaz', 'Waleed khan', 'Asim khalil'],
'age': [20, 21, 22, 23],
'location': ['karachi pakistan', 'lahore USA', 'islamabad UK', 'peshawar Sweden']})

df
```

```python
#spliting two columns

# df.name.str.split(' ', expand=True).head()
df.location.str.split(' ', expand=True).head()
# df
```

```python
# adding those splits into new columns
# df['first_name'] = df.name.str.split(' ', expand=True)[0]
df[['first_name','last_name']] =df.name.str.split(' ', expand=True).head()
df[['city', 'country']] =  df.location.str.split(' ', expand=True).head()
df
```

```python
df = df[['first_name', 'last_name' , 'city', 'country']]
df
```
