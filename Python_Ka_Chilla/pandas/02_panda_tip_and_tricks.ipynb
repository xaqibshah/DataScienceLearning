## 03- How to rename  columns?

```python
df = pd.DataFrame({'A col': [1,2,3,4,10,25,35,50,40,98], 'B col':[5,6,7,8,101,151,19,98,105,110]})
#Remane column
df.rename(columns={"A col": 'col_a', 'B col': 'col_b'}, inplace=True)
df
```

```python
#another way to change column name
df.columns = ['col_aa', 'col_bb']
df
```

```python
# to replace any character, string
df.columns = df.columns.str.replace('_', '-')
df
```

```python
df.columns = df.columns.str.replace('-', '*')
df
```

```python
df = df.add_prefix('**')
```

```python
df.add_suffix('**')
```

```python
df.columns = ['col-a', 'col-b']
df
```
